```python
from bs4 import BeautifulSoup
import requests
import pandas
```

```python
#get the url page to scrap
urlPage = "https://latitude.to/map/cn/china/cities/by-population/page/"
urlLoc = "https://latitude.to"
```

```python
def getCities(url, page):
    r = requests.get(url+str(page))
    soup = BeautifulSoup(r.content, 'html.parser')
    citiesTab = soup.find("div", {"class":"b-cities-listing cols-3"}).find("ul")
    
    cities_a = citiesTab.find_all("a")
    
    return cities_a

def getCitiesLoc(url):
    r = requests.get(urlLoc + url)
    soup = BeautifulSoup(r.content, 'html.parser')
    lat = soup.find("meta", {"property":"place:location:latitude"})["content"]
    long = soup.find("meta", {"property":"place:location:longitude"})["content"]
    return [lat, long]
    
```

```python
cities_a_list = []
for i in range(1,26):
    cities_a_list += getCities(urlPage, i)
print(len(cities_a_list))
```

```python
cities = {}
nbCities = len(cities_a_list)
for a in cities_a_list:
    cities[a.text] = getCitiesLoc(a['href'])
    print(str(len(cities))+"/" + str(nbCities))

print(cities)
```
